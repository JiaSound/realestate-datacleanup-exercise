#### Exercise 01. Which is the most expensive house in the dataset? (★☆☆)

Print the address and the price of the selected house. For example:

`The house with address General Street Nº5 is the most expensive and its price is 5000000 USD`

```python
# TODO


df = pd.DataFrame(ds) # Converting DS Raw data to a Data Frame
#df.info() #see all columns and data types

maxx = df['price'].max() # Finding the highest cost house
full_row = df[df['price'] == maxx] #pulling the row with the value that matches maxx (highest value house)
addy1 = full_row['address'].iloc[0] #pulling address from the isolated row
print(f' The house with address {addy1} is the most expensive and its price is {maxx} USD')

#------------------------------------------------------------------------------------------------

#Cleaner way to do this:

most_expensive_house = df.loc[df['price'].idxmax()] # Extract the most expensive house's entire row in one line with idxmax
print(f'The house with address {most_expensive_house["address"]} is the most expensive and its price is {most_expensive_house["price"]} USD')
```

Print both average prices and then write a conclusion about them.

```python
# TODO

val = df[df['level5'] == 'Valdemorillo']
gal = df[df['level5'] == 'Galapagar']
v_mean = val['price'].mean()
g_mean = gal['price'].mean()


answer = ""
if v_mean == g_mean:
    answer = "They are both equal"
else:
    answer = "They're not equal"

print(f'''Valdemorillo Mean: {float(v_mean):.2f}
Galapagar Mean: {float(g_mean):.2f}
-------------------
{answer}
-------------------''')

```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO

val_pps_list = []
gal_pps_list = []

for index, i in val['price'].items():
    val_pps_list.append(float(round(i/val.loc[index, 'surface'],2)))
val['pps'] = val_pps_list


for index, i in gal['price'].items():
    gal_pps_list.append(float(round(i/gal.loc[index, 'surface'],2)))
gal['pps'] = gal_pps_list


val_pps_av = round(float(val['pps'].mean()))
gal_pps_av = round(float(gal['pps'].mean()))


answer = ""

if val_pps_av == gal_pps_av:
    answer = "They are both equal"
else:
    answer = "They're not equal"

print(f'''


Valdemorillo Average PPS: {float(val_pps_av):.2f}
Galapagar Average PPS: {float(gal_pps_av):.2f}
-------------------
{answer}
-------------------


I am ignoring the warning below on purpose, 
I am aware that I am overwriting the column I created


''')
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)
